Now, I will the data from the [List of helicopter prison escapes](https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes) Wikipedia article.

```python
index = 0
for row in data:
    data[index] = row[:-1]
    index +=1
print(data[:2])
```

Now, I would like to change the date values to include just the Year value.

```python
for row in data:
    row[0] = fetch_year(row[0])
print(data[:2])
```

Now I want to determine the minimum and maximum years in the dataset.

```python
min_year = min(data, key=lambda x: x[0])[0]
max_year = max(data, key=lambda x: x[0])[0]
```

```python
years = []
for year in range(min_year, max_year + 1):
    years.append(year)
```

```python
print(years)
```

```python
attempts_per_year = []
for i in years:
    attempts_per_year.append([i,0])
```

```python
print(attempts_per_year)
```

```python
for row in data:
    for year_attempt in attempts_per_year:
        year = year_attempt[0]
        if row[0] == year:
            year_attempt[1] += 1
            
print(attempts_per_year)   
```
